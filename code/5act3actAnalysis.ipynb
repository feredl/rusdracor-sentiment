```python
import pandas as pd
import plotly.express as px
```

```python
df5 = pd.read_excel("5XLMResCount.xlsx")
```

```python
df5
```

```python
comedy_act_data = df5[df5['normalizedGenre'] == 'Comedy']
tragedy_act_data = df5[df5['normalizedGenre'] == 'Tragedy']
ns_act_data = df5[df5['normalizedGenre'] == 'NotStated']
```

```python
percent_columns = [col for col in comedy_act_data.columns if col.startswith('percent')]
comedy_act_data = comedy_act_data.groupby('act')[percent_columns].mean().reset_index()
tragedy_act_data = tragedy_act_data.groupby('act')[percent_columns].mean().reset_index()
ns_act_data = ns_act_data.groupby('act')[percent_columns].mean().reset_index()
```

```python
comedy_act_data
```

```python
import plotly.express as px

data_frames = [
    (comedy_act_data, "комедиях"),
    (tragedy_act_data, "трагедиях"),
    (ns_act_data, "пьесах без жанра")
]

rename_dict = {
    "percentPOSITIVE": "POSITIVE",
    "percentNEGATIVE": "NEGATIVE",
    "percentNEUTRAL": "NEUTRAL",
    "percentSKIP": "SKIP",
    "percentSPEECH": "SPEECH"
}

html_plots = []

for df, genre in data_frames:
    decade_fig = px.line(
        df,
        x="act",
        y=list(rename_dict.keys()),
        title=f'Соотношение количества реплик разнвх тональностей русской драмы XVIII-XX вв. по актам в {genre}',
        labels={"value": "%"},
        color_discrete_map={
            "percentPOSITIVE": "limegreen",
            "percentNEGATIVE": "crimson",
            "percentNEUTRAL": "royalblue",
            "percentSKIP": "plum",
            "percentSPEECH": "grey"
        }
    )

    decade_fig.for_each_trace(lambda t: t.update(name=rename_dict[t.name]))

    decade_fig.update_layout(
        title_x=0.5,
        font_family="Bahnschrift",
        legend_title_text='Тональность',
        template="simple_white",
        legend=dict(
            yanchor="top",
            y=0.99,
            xanchor="left",
            x=0.01
        ),
        hovermode='x unified'
    )

    decade_fig.update_traces(
        line={'width': 5},
        hovertemplate="%{y:.2f}%<extra></extra>"
    )

    decade_fig.update_xaxes(dtick=1, range=[1, 5])

    decade_fig.update_yaxes(showgrid=True, gridwidth=1, gridcolor='LightGray')

    html_plots.append(decade_fig.to_html(full_html=False))

    decade_fig.show()

    with open("all_plots.html", "w", encoding="utf-8") as f:
      f.write("<html><head><meta charset='UTF-8'><title>Графики</title></head><body>")
      for html_plot in html_plots:
          f.write(html_plot)
      f.write("</body></html>")
```
